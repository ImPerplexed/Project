```python
Company_Info = {
            "Name": [],
            "Industry": [],
            "Employee": [],
            "Revenue": [],
            "Overview": [],
            "Website": [],
            "Headquarters": [],
            "Company_Type": [],
            "Founded_Year": [],
            "Speciality": []}
Company_Info_Sum = {
            "Company_Name": [],
            "Company_Link": []}
```

```python
# direct the webdriver to where the browser file is:
driver_path = r"C:\Users\Kiet\geckodriver.exe"
driver = webdriver.Firefox()
# your secret credentials:
email = "vho@jkadvisory.us"
password = "Vietnam2022"
# Go to linkedin and login
driver.get('https://www.linkedin.com/login')
time.sleep(3)
driver.find_element(By.ID, "username").send_keys(email)
driver.find_element(By.ID, "password").send_keys(password)
driver.find_element(By.ID, "password").send_keys(Keys.RETURN)
time.sleep(6)
```

```python
WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.XPATH,'//*[@id="search-results-container"]')))
    time.sleep(8)
    inner_div = driver.find_element(By.XPATH, '//*[@id="search-results-container"]')
    for k in range(9):
        inner_div.send_keys(Keys.PAGE_DOWN)
        time.sleep(1)
    list_items = driver.find_elements(By.XPATH, "/html/body/main/div[1]/div[2]/div[2]/div[2]/ol/li/div/div/div[2]/div/div/div/div[2]/div/div/a")
    Company_Name = [item.text for item in list_items]
    Company_Link = [item.get_attribute('href') for item in list_items]
    # print(Company_Name)
    [Company_Info_Sum['Company_Name'].append(i) for i in Company_Name]
    [Company_Info_Sum['Company_Link'].append(i) for i in Company_Link]
    
    driver.execute_script("window.open('');") 
    driver.switch_to.window(driver.window_handles[1])
    for item in Company_Link:
        driver.get(item)
        time.sleep(4)
        try:
            driver.find_element(By.XPATH, "/html/body/main/section/section/div[2]/section[1]/div[1]/div[2]/p/div/span[2]/button")
        except:
            Company_Info["Name"].append(np.NaN)
            Company_Info["Industry"].append(np.NaN)
            Company_Info["Employee"].append(np.NaN)
            Company_Info["Revenue"].append(np.NaN)
            Company_Info["Overview"].append(np.NaN)
            Company_Info["Website"].append(np.NaN)
            Company_Info["Headquarters"].append(np.NaN)
            Company_Info["Company_Type"].append(np.NaN)
            Company_Info["Founded_Year"].append(np.NaN)
            Company_Info["Speciality"].append(np.NaN)
            continue
        else
        Name = driver.find_element(By.XPATH,
                                   "/html/body/main/section/section/div[2]/section[1]/div[1]/div[1]/div/div[2]/div").text
        Industry = driver.find_element(By.XPATH,
                                               "/html/body/main/section/section/div[2]/section[1]/div[1]/div[1]/div/div[2]/div[2]/div/span[1]").text
        try:
            Employee = driver.find_element(By.XPATH,
                                   "/html/body/main/section/section/div[2]/section[1]/div[1]/div[1]/div/div[2]/div[2]/div/span[3]").text
        except:
            Employee = np.NaN
    
        try:
            Revenue = driver.find_element(By.XPATH,
                            "/html/body/main/section/section/div[2]/section[1]/div[1]/div[1]/div/div[2]/div[2]/div/span[4]/span[2]").text
        except:
            try:
                Revenue = driver.find_element(By.XPATH,
                                      "/html/body/main/section/section/div[2]/section[1]/div[1]/div[1]/div/div[2]/div[2]/div/span[2]").text
            except:
                Revenue = np.NaN
                
        more_info =driver.find_element(By.XPATH, "/html/body/main/section/section/div[2]/section[1]/div[1]/div[2]/p/div/span[2]/button").click()
        time.sleep(1)
        try:
            driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div[2]/div/dl[1]/dd/a/span").text
        except:
            continue
            
        Overview = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div[2]/p[2]").text
        if np.count_nonzero(driver.find_elements(By.XPATH, "/html/body/div[7]/div/div/div[2]/div//dl")) == 4:
            Website = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div[2]/div/dl[1]/dd/a/span").text
            Headquarters = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div[2]/div/dl[2]/dd").text
            Company_Type = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div[2]/div/dl[3]/dd").text
            Founded_Year = np.NaN
            Speciality = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div[2]/div/dl[4]/dd").text
        elif np.count_nonzero(driver.find_elements(By.XPATH, "/html/body/div[7]/div/div/div[2]/div//dl")) == 5:
            Website = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div[2]/div/dl[1]/dd/a/span").text
            Headquarters = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div[2]/div/dl[2]/dd").text
            Company_Type = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div[2]/div/dl[3]/dd").text
            Founded_Year = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div[2]/div/dl[4]/dd").text
            Speciality = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div[2]/div/dl[5]/dd").text
        time.sleep(1)
        
        webdriver.ActionChains(driver).send_keys(Keys.ESCAPE).perform()
        
        Company_Info["Name"].append(Name)
        Company_Info["Industry"].append(Industry)
        Company_Info["Employee"].append(Employee)
        Company_Info["Revenue"].append(Revenue)
        Company_Info["Overview"].append(Overview)
        Company_Info["Website"].append(Website)
        Company_Info["Headquarters"].append(Headquarters)
        Company_Info["Company_Type"].append(Company_Type)
        Company_Info["Founded_Year"].append(Founded_Year)
        Company_Info["Speciality"].append(Speciality)
    
        # print(Name, "/", Industry, "/", Employee, "/", Revenue, "/", Website, "/", 
        #       Headquarters, "/", Founded_Year, "/", Company_Type, "/", Speciality)
        
        for i in range(np.random.randint(1, 6)):
            webdriver.ActionChains(driver).send_keys(Keys.PAGE_DOWN).perform()
            time.sleep(0.5)
        for i in range(np.random.randint(1, 6)):
            webdriver.ActionChains(driver).send_keys(Keys.PAGE_UP).perform()
            time.sleep(0.5)
        time.sleep(0.5)
        Company_Info_DataFrame = pd.DataFrame(Company_Info)
        Company_Info_DataFrame.to_csv("CompanyInfo_BussinessConsultingServices_US_Page1-5")
        Company_Info_Sum_DataFrame = pd.DataFrame(Company_Info_Sum)
        Company_Info_Sum_DataFrame.to_csv("CompanyInfo_Sum_BussinessConsultingServices_US_Page1-5")
        display(Company_Info_DataFrame)
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    driver.find_element(By.XPATH, "/html/body/main/div[1]/div/div[2]/div[2]/div[5]/div/button[2]/span").click()
```
