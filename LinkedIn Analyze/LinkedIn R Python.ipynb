```python
import pandas as pd

df1 = pd.read_excel('linked1_3.xlsx', index_col=0, sheet_name='Sheet1')
df2 =  pd.read_excel('linked4_9.xlsx', index_col=0, sheet_name='Sheet1')
df = pd.concat([df1, df2])
df = df.drop_duplicates(subset = ['Job', 'Company', 'Detail'])
df = df.reset_index(drop = True)

df['Python'] = df['Detail'].str.contains('Python', regex = False)
df['R'] = df['Detail'].str.contains('\sR[,.\s]', regex = True)
display(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette("pastel", as_cmap=False)
plt.figure(figsize = (16,9))
plt.pie([df['Python'].mean(), df['R'].mean(), 1 - df['Python'].mean() - df['R'].mean() ], labels = ['Python', 'R', 'Not mentioned'], 
        startangle = 90, autopct='%1.1f%%', colors = color, textprops={'fontsize': 16})
plt.title('Python and R Requirement \non LinkedIn Job Description for Data Analyst', fontsize = 16)
plt.show()
```
